1. What is the difference between a function and a method in Python?

A **function** is defined outside a class, and a **method** is defined inside a class and works on objects.
Example:
Function → `len([1,2,3])`
Method → `"hello".upper()`

2. Explain the concept of function arguments and parameters in Python.

**Parameters** are the names in the function definition, and **arguments** are the actual values passed.
Example:
`def add(a,b):` → `a,b` = parameters
`add(3,5)` → `3,5` = arguments

3. What are the different ways to define and call a function in Python?

Functions can be defined using `def` or `lambda`.
They can be called by writing the function name followed by parentheses.
Example:
`def fun(): print("hi")`
`fun()`
Lambda → `(lambda x: x+1)(5)`

4. What is the purpose of the `return` statement in a Python function?

`return` gives the output of a function and ends the function.
Example:
`def add(a,b): return a+b`

5. What are iterators in Python and how do they differ from iterables?

An **iterator** can be used with `next()` and remembers its state; an **iterable** can be looped over but does not remember state.
Example:
`iter([1,2,3])` → iterator
`[1,2,3]` → iterable

6. Explain the concept of generators in Python and how they are defined.

Generators are functions that use `yield` to produce values one at a time.
Example:
`def gen(): yield 1; yield 2`

7. What are the advantages of using generators over regular functions?

Generators use less memory and are faster for large data because they produce values one by one.
Example:
Iterating over `range(10**9)` (generator) uses almost no memory.

8. What is a lambda function in Python and when is it typically used?

A lambda is a small anonymous function used for short tasks.
Example:
`lambda x: x*x`

9. Explain the purpose and usage of the `map()` function in Python.

`map()` applies a function to each item of a sequence.
Example:
`map(str, [1,2,3])` → `["1","2","3"]`

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

`map()` transforms items, `filter()` selects items, and `reduce()` combines items into one value.
Example:
Map → `map(lambda x:x+1,[1,2,3])`
Filter → `filter(lambda x:x%2==0,[1,2,3])`
Reduce → sums list into one number
